```python
df_96_11 = []
for i in range(1996, 2012):
    df = pd.read_csv('Datasets/datos_'+str(i)+'.csv', encoding='latin1')
    df = df[df['PARAMETRO']=='CO']
    df = df[df.applymap(np.isreal).any(1)]
    df['FECHA'] = pd.to_datetime(df['FECHA'], format="%d/%m/%y")
    df['PROM'] = df.median(axis=1)
    df.index = df['FECHA']
    df_96_11.append(df)
df_96_11 = pd.concat(df_96_11, ignore_index=True)
```

## Traer datos 2016-2017

```python
def prom_mensual(anio, fecha_label):
    df = pd.read_csv('Datasets/datos_'+anio+'.csv', encoding='latin1', index_col=fecha_label)
    df.index = pd.to_datetime(df.index)
    df = df[df['CO']>0.25]
    df = df[df['CO']<3]
    r = df.resample('d').median()
    return r
```

```python
df16 = prom_mensual('2016', 'Fecha')
df15 = prom_mensual('2015', 'Fecha')
df_15_16 = pd.concat([df15, df16])
```
