```python
df_96_11 = []
for i in range(1996, 2012):
    df = pd.read_csv('Datasets/datos_'+str(i)+'.csv', encoding='latin1')
    df = df[df['PARAMETRO']=='CO']
    df['FECHA'] = pd.to_datetime(df['FECHA'], format="%d/%m/%y")
    df['PROM'] = df.mean(axis=1)
    df.index = df['FECHA']
    df_96_11.append(df)
```

```python
df_96_11 = pd.concat(df_96_11, ignore_index=True)
```

```python
plt.plot(df_96_11['FECHA'], df_96_11['PROM'], 'bo')
```

```python
df_96_11.index = df_96_11['FECHA']
```

```python
df_96_11_month_median = df_96_11.resample('M').median()
```

```python
df_96_11_month_median
```

```python
plt.figure(figsize=(16,9))

plt.plot(df_96_11_month_median['PROM'].dropna())
```

```python
pm10_96_11 = []
for i in range(2000, 2012):
    df = pd.read_csv('Datasets/datos_'+str(i)+'.csv', encoding='latin1')
    df = df[df['PARAMETRO']=='CO']
    df['FECHA'] = pd.to_datetime(df['FECHA'], format="%d/%m/%y")
    df['PROM'] = df.mean(axis=1)
    df.index = df['FECHA']
    pm10_96_11.append(df)
pm10_96_11 = pd.concat(pm10_96_11, ignore_index=True)
pm10_96_11.index = pm10_96_11['FECHA']
```

```python
pm10_96_11_month_median = pm10_96_11.resample('M').mean()
```

```python
plt.figure(figsize=(16,9))

plt.scatter(pm10_96_11_month_median.index, pm10_96_11_month_median['PROM'])
plt.plot(pm10_96_11_month_median.index, pm10_96_11_month_median['PROM'], 'b-')
```
