```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def prom_mensual(anio, fecha_label):
    df = pd.read_csv('Datasets/datos_'+anio+'.csv', encoding='latin1', index_col=fecha_label)
    df.index = pd.to_datetime(df.index)
    r = df.resample('M').median()
    return r
```

```python
df_2016 = prom_mensual('2016','Fecha')
df_2015 = prom_mensual('2015', 'Fecha')
plt.plot(df_2016['CO'])
plt.plot(df_2015['CO'])
```
